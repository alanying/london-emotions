```python
import pandas as pd
text = 'Hello world, I am feeling great today! The world is wonderful YES!!!'
text_df = {
    'Text': [text]
}
text_df = pd.DataFrame(text_df)
text_df.head()
```

```python
from LondonEmotions.data import clean_data
clean_text = clean_data(text_df)
clean_text
```

```python
text_value = clean_text['tokenized_text']
text_value[0]
```

```python
# save and load tokenizer: 
#https://stackoverflow.com/questions/45735070/keras-text-preprocessing-saving-tokenizer-object-to-file-for-scoring
#
# GCP path: gs://wagon-ml-london-emotions/models/tokenizer/model_tokenizer
from tensorflow.python.lib.io import file_io
import pickle
# with file_io.FileIO('models_tokenizer_model_tokenizer.pickle', 'rb') as handle:
# #     tokenizer = pickle.load(handle)
# tokenizer = pickle.load('gs://wagon-ml-london-emotions/models/tokenizer/model_tokenizer.pickle')

#local?
# tokenizer = models_tokenizer_model_tokenizer.pickle
with file_io.FileIO('../models_tokenizer_model_tokenizer', 'rb') as handle:
     tokenizer = pickle.load(handle)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
max_seq_len = 300
index_of_words = tokenizer.word_index
vocab_size = len(index_of_words) + 1
sequence_text = tokenizer.texts_to_sequences(text_value)
text_pad = pad_sequences(sequence_text, maxlen=max_seq_len)
text_pad
```

```python
from tensorflow.keras.models import load_model
model = load_model('../raw_data/saved_model.pb')
preds = model.predict(text_pad)
```
